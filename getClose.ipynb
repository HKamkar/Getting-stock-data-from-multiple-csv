```python
# import OS module
import os
import csv
from pathlib import Path
import pandas as pd
from pandas import isna
 
# Get the list of all files and directories
data_folder = Path(r'C:\Users\HKamkar\OneDrive\Desktop\Hossein Kamkar\data')
dir_list = os.listdir(data_folder)
 
print("Files and directories in '", data_folder, "' :")
 
# prints all files
print(dir_list)
```

```python
print(str(data_folder) + '\AKHABER.csv')
```

```python
df1 = pd.read_csv(str(data_folder) + '\ALBORZ.csv', usecols= ['<DTYYYYMMDD>','<CLOSE>'])
df1.set_index('<DTYYYYMMDD>')
df1.rename(columns={'<Close>': 'ALBORZ'})
```

```python
df2 = pd.read_csv(str(data_folder) + '\ASIA.csv', usecols= ['<DTYYYYMMDD>','<CLOSE>'])
df2.set_index('<DTYYYYMMDD>')
df2.rename(columns={'<Close>': 'ASIA'})
```

```python
frames = [df1, df2]
result = pd.merge(df1, df2, how="left", on='<DTYYYYMMDD>')
result
```

```python
data = pd.DataFrame(df1)
data.drop(columns='<CLOSE>', inplace = True)
df = pd.DataFrame()
data
```

```python
for stock_data in dir_list:
    #print(stock_data)
    df = pd.read_csv(str(data_folder) + '\\' + stock_data, usecols= ['<DTYYYYMMDD>','<CLOSE>'])
    df = df.set_index('<DTYYYYMMDD>')
    #print(stock_data)
    stock_name = stock_data.split('.')
    #print(stock_name)
    df.rename(columns={'<CLOSE>': stock_name[0].casefold().capitalize() }, inplace = True)
    data = pd.merge(data, df, how="outer", on='<DTYYYYMMDD>')
```

```python
data = data.sort_values(by=['<DTYYYYMMDD>'])
data = data.set_index('<DTYYYYMMDD>')
data
```

```python
# Number of missing Values for each individual stock
data.isna().sum()
```

```python
# How many stocks need back filling (na in first rows)
data.iloc[0,:].isna().sum()
```

```python
# Filling the missing data
data = data.fillna(method='ffill')
data = data.fillna(method='bfill')
data
```

```python
# Filling the missing data
# Drop holidays (non trading days)
data.dropna(axis = 0, how = 'all', inplace = True)
```

```python
# Number of missing Values for each individual stock
data.isna().sum()
```

```python
data.to_csv('close.csv', index=True)
```
