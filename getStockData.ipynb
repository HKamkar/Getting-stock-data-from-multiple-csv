```python
# import OS module
import os
import csv
from pathlib import Path
import pandas as pd
from pandas import isna
import numpy as np
 
# Get the list of all files and directories
data_folder = Path(r'C:\Users\HKamkar\OneDrive\Desktop\Hossein Kamkar\data')
dir_list = os.listdir(data_folder)
 
print("Files and directories in '", data_folder, "' :")
 
# prints all files
print(dir_list)
```

```python
company_list = []

for stock_data in dir_list:
    stock_name = stock_data.split('.')
    company_list.append(stock_name[0])
```

```python
features= ['<DTYYYYMMDD>','<OPEN>','<HIGH>','<LOW>','<CLOSE>','<VOL>']
data = pd.DataFrame(columns=features)

for stock_data in dir_list:
    #print(stock_data)
    df = pd.read_csv(str(data_folder) + '\\' + stock_data, usecols= features)
    #df = df.set_index('<DTYYYYMMDD>')
    stock_name = stock_data.split('.')
    stock_name = np.array(stock_name[0])
    #print(stock_name)
    df['Name'] = stock_name
    data = pd.concat([data,df])
```
